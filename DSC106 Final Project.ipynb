```python
def latcoords(loc):
    lat = {
     'Back Bay': 42.3495,
     'Beacon Hill':42.3562,
     'Boston University':42.3505,
     'Fenway':42.3467,
     'Financial District':40.7077,
     'Haymarket Square':42.3638,
     'North End':42.3652,
     'North Station':42.3664,
     'Northeastern University':42.3399,
     'South Station':42.3519,
     'Theatre District':40.7590,
     'West End':26.6867
    }
    

    return lat[loc]

def longcoords(loc):
        long = {
         'Back Bay': -71.0795,
         'Beacon Hill':-71.0694,
         'Boston University':-71.1054,
         'Fenway':-71.0972,
         'Financial District':-74.0083,
         'Haymarket Square':-71.0585,
         'North End':-71.0555,
         'North Station':-71.0620,
         'Northeastern University': -71.0899,
         'South Station':-71.0552,
         'Theatre District':-73.9845,
         'West End':-78.9712

        }
        
        return long[loc]
```

```python
uber_lyft['pickup_latitude'] = uber_lyft['source'].apply(latcoords)
uber_lyft['pickup_longitude'] = uber_lyft['source'].apply(longcoords)
uber_lyft['dropoff_latitude'] = uber_lyft['destination'].apply(latcoords)
uber_lyft['dropoff_longitude'] = uber_lyft['destination'].apply(longcoords)
```

```python
import pandas as pd

# Read the weather CSV into a DataFrame
weather_df = pd.read_csv("weather.csv")

# Convert the timestamp column to datetime format
weather_df["time_stamp"] = pd.to_datetime(weather_df["time_stamp"], unit="s")

# Extract only the date from the timestamp
weather_df["date"] = weather_df["time_stamp"].dt.date

weather_df
```
